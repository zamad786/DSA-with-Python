```python
b = [9,8,7,6,4,3,2]
c = b[0]
for i in b:
  if i < c:
    c = i
print("Lowest value: ", c)
```

```python
a = [2,3,4,9,8,7,6,5]
b = a[0]
for i in a:
  if i > b:
    b = i
print("Less value: ", b)


if 7 in a:
  print("Yes 7 is in the list.")
else:
  print("No there is no 7 in the list")

a.append(12)
print(a)

a.remove(6)
print(a)
```
