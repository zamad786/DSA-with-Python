```python
a = []
#push
a.append("A")
a.append("B")
a.append("C")
a.append("D")
print(a)
a.append(1)
a.append(2)
a.append(3)
a.append(4)
print(a)

#peek
b = a[0]
print("Bottom element: ", b)
b = a[2]
print("upper element: ", b)
b = a[-1]
print("Top element: ", b)
b = a[-3]
print("Les Top element: ", b)

#pop
c = a.pop()
print("Pop: ", c)
c = a.pop(0)
print("Pop: ", c)
c = a.pop(2)
print("Pop: ", c)
c = a.pop(-2)
print("Pop: ", c)

#stack after pop
print("Pop: ", a)

#is empty
emty = not bool(a)
print("Checking Stack: ", emty)

ty = bool(a)
print("check: ", ty)

#size
print("Size: ", len(a))
```
